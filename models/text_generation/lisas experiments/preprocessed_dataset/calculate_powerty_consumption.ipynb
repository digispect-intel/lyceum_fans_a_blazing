```python
import pandas as pd

input_file = 'merged_runtime_results.xlsx'

df = pd.read_excel(input_file)

power_cpu = 20     # Ватты для CPU (примерное)
power_gpu = 100    # Ватты для GPU (примерное)

def estimate_power(row):
    runtime_hours = row['runtime_sec'] / 3600
    if str(row['device']).lower() == 'gpu':
        return power_gpu * runtime_hours
    else:
        return power_cpu * runtime_hours

df['powerty_consumption'] = df.apply(estimate_power, axis=1)

df.to_excel(input_file, index=False)

print(f"Столбец 'estimated_energy_Wh' добавлен и сохранён в {input_file}")
```
