```python
import os
import pandas as pd

# Путь к твоим папкам (убедись, что путь верный)
cpu_folder = "/content/drive/MyDrive/dataset/cpu"
gpu_folder = "/content/drive/MyDrive/dataset/gpu"


all_folders = [cpu_folder, gpu_folder]

# Чтение всех Excel-файлов
all_dfs = []

for folder in all_folders:
    for filename in os.listdir(folder):
        if filename.endswith(".xlsx"):
            file_path = os.path.join(folder, filename)
            try:
                df = pd.read_excel(file_path)
                df['source_file'] = filename
                df['source_device'] = os.path.basename(folder)
                all_dfs.append(df)
                print(f"Добавлен: {filename}")
            except Exception as e:
                print(f"Ошибка при чтении {filename}: {e}")

# Объединение
if all_dfs:
    final_df = pd.concat(all_dfs, ignore_index=True)
    output_path = os.path.join( "merged_runtime_results.xlsx")
    final_df.to_excel(output_path, index=False)
    print(f"\nОбъединённый файл сохранён как: {output_path}")
else:
    print("Нет данных для объединения.")
```
