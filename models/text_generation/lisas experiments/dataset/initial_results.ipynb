```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Загрузка данных из файла (замени путь и формат, если нужно)
df_test = pd.read_excel('test_predictions.xlsx')  # или pd.read_csv('test_predictions.csv')


def plot_predictions_vs_actual(y_true, y_pred, target_name):
    plt.figure(figsize=(12,5))

    # Scatter plot реальные vs предсказанные
    plt.subplot(1,2,1)
    plt.scatter(y_true, y_pred, alpha=0.6)
    plt.plot([y_true.min(), y_true.max()], [y_true.min(), y_true.max()], 'r--')  # линия y=x
    plt.xlabel('Actual ' + target_name)
    plt.ylabel('Predicted ' + target_name)
    plt.title(f'Actual vs Predicted for {target_name}')
    plt.grid(True)

    # Гистограмма ошибок
    plt.subplot(1,2,2)
    errors = y_pred - y_true
    sns.histplot(errors, bins=30, kde=True)
    plt.title(f'Error distribution for {target_name}')
    plt.xlabel('Prediction Error')
    plt.grid(True)

    plt.tight_layout()
    plt.show()


plot_predictions_vs_actual(df_test['runtime_actual'], df_test['runtime_pred'], 'runtime')
plot_predictions_vs_actual(df_test['power_actual'], df_test['power_pred'], 'powerty_consumption')
```
